COVID Observation (2020–2021)

Observation:
The number of titles added on Netflix decreased in 2020 and 2021 compared to 2019.
Possible Reason:
This decline may have been influenced by the COVID-19 pandemic. During the pandemic, film and television productions around the world were delayed or temporarily halted due to lockdowns and safety restrictions. As a result, fewer new titles may have been available for Netflix to add to its catalog.

International Movies Dominating in India

Observation:
International Movies were the most common category in the Netflix dataset.
Possible Reason:
Netflix is a global streaming platform that serves audiences across many countries. To attract and retain subscribers worldwide, Netflix offers content from different regions and cultures rather than focusing on only one country. Therefore, a large number of international movies are included in its catalog, making International Movies the most common category in the dataset.

Country Analysis

Observation:
The United States had the highest number of titles in the dataset, followed by India.
Possible Reason:
Netflix originated in the United States and has strong partnerships with American studios and production houses. India appears in second place because of its large entertainment industry and growing streaming audience.
